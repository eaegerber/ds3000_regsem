```python
# Solution

import numpy as np
import pandas as pd
import seaborn as sns

# this can control the maximum number of rows displayed in a data frame without having to call a custom .head()
pd.options.display.max_rows = 500

rename_dict = {'body_mass_g': 'weight (g)',
              'bill_depth_mm': 'beak depth (mm)',
              'bill_length_mm': 'beak length (mm)',
              'flipper_length_mm': 'flipper (mm)'}

df_penguins = sns.load_dataset('penguins')
df_penguins.sort_values(['island', 'species', 'body_mass_g'], axis=0, inplace=True)
df_penguins.set_index(['island', 'species'], inplace=True)
df_penguins.sort_index(axis=1, inplace=True)
df_penguins.rename(rename_dict, axis=1, inplace=True)
df_penguins
```

# Lecture Break/Practice 2 (Grades)

```python
# solution
grades_df = pd.read_csv('grades.csv', na_values=(-99))

grades_df
```

```python
# two lines below are equivilent
# grades_df.fillna(0, inplace=True)
grades_df.replace(np.nan, 0, inplace=True)

grades_df
```

```python
column = grades_df.loc[:, 'quiz0']
column.sort_values()[1:-1].mean()
```

```python
def mean_skip_minmax(column):
    return column.sort_values()[1:-1].mean()
```

```python
grades_ind = grades_df.set_index('name')
grades_ind.apply(mean_skip_minmax, axis=0)
```

```python
# how to do it without setting the index?
grades_df.loc[:, 'quiz0':].apply(mean_skip_minmax, axis=0)
```
